# Simple introduction to matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 20, 100)
```

```python
y = np.exp(-0.3*t) * np.sin(2*t)
```

**A line chart or line graph is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.**

### https://en.wikipedia.org/wiki/Line_chart

```python
plt.title("signal")
plt.ylabel("amplitude")
plt.xlabel("time")
plt.plot(t, y)
```

```python
plt.title("signal")
plt.ylabel("amplitude")
plt.xlabel("time")

# customize color "r" (for red), marker '^' and line style '-.'
plt.plot(t, y, "r^-.")
```

```python
plt.title("signal")
plt.ylabel("amplitude")
plt.xlabel("time")

t = np.linspace(0, 5, 100)

# more than one line graph
# x1, y1, x2, y2, x3, y3....

plt.plot(t, t**2, t, 0.5 * t**3, t, 10 * np.sin(2*t))
```

```python
# x for scatter plot

x = np.linspace(0, 5, 25)

# generate Gaussian noise and add it to x**2

y = x**2 + np.random.randn(25)
```

**Scatter plot is a type of plot or mathematical diagram using Cartesian coordinates
to display values for typically two variables for a set of data.**

### https://en.wikipedia.org/wiki/Scatter_plot

```python
plt.ylabel("y(x)")
plt.xlabel("x")

plt.scatter(x, y)
```

```python
plt.ylabel("y(x)")
plt.xlabel("x")

# if you need to just plot points without much of customization of individual points (size, color etc.),
# it is better to use "plot" with "empty line style". That is faster than "scatter" for large dataset

plt.plot(x, y, marker="o", linestyle=' ')
```

**Boxplot is a method for graphically depicting groups of numerical data through their quartiles.**

### https://en.wikipedia.org/wiki/Box_plot ###

```python
plt.boxplot(y)
```

```python
plt.boxplot(y, vert=False)
```

**A contour plot is a graphical technique for representing a 3-dimensional surface by plotting constant z slices,
called contours, on a 2-dimensional format**

"Countour lines are lines of equal value in maps and diagrams"

https://en.wikipedia.org/wiki/Contour_line

https://www.itl.nist.gov/div898/handbook/eda/section3/contour.htm

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
```

```python
# https://www.quora.com/Can-you-explain-to-me-what-the-meshgrid-function-in-Python-does-step-by-step
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html

# meshgrid returns coordinate matrices from coordinate vectors. 
X, Y = np.meshgrid(x, y)

# calculate Z values for each point in mesh grid
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
```

```python
plt.contour(Z)
```

**Countourf draws contour lines and filled contours**

```python
# contour"f" = "f" for filled

plt.contourf(Z)

# show contourbar for the legend for colors
plt.colorbar()
```

**Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.**

### https://seaborn.pydata.org/

```python
import seaborn as sns
```

**Iris flower data set or Fisher's Iris data set is a data set introduced by the British statistician and biologist Ronald Fisher**

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

### https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
# load "iris" dataset that comes with seaborn
iris = sns.load_dataset("iris")

type(iris)
```

```python
# descriptive info about iris DataFrame
iris.info()
```

```python
iris.head()
```

```python
# DataFrame's plot method

iris.plot()
```

```python
# Plot pairwise relationships in a dataset.
sns.pairplot(iris)
```

```python
# Plot pairwise relationships in a dataset. "hue" variable in data to map plot aspects to different colors.
# In this example, we use "species" as "hue" so that different flower species data will be shown in different colors.

sns.pairplot(iris, hue="species")
```
